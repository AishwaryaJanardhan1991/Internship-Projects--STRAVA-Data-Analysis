```python
# Converting 'ActivityDay' to datetime
daily_steps['ActivityDay'] = pd.to_datetime(daily_steps['ActivityDay'])

# Creating a new column for weekday names
daily_steps['Weekday'] = daily_steps['ActivityDay'].dt.day_name()

# Group by Id and Weekday to get average steps
grouped_data = daily_steps.groupby(['Id', 'Weekday'])['StepTotal'].mean().reset_index()

# Pivot for heatmap
heatmap_data = grouped_data.pivot(index='Id', columns='Weekday', values='StepTotal')

# Reorder the columns to match Sunday to Saturday
ordered_days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
heatmap_data = heatmap_data[ordered_days]

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd', linewidths=0.5, linecolor='Black')
plt.title('Avg. Steps Taken by Participants on Weekdays', fontsize=16)
plt.xlabel('Weekday')
plt.ylabel('User ID')
plt.tight_layout()
plt.show()
```
